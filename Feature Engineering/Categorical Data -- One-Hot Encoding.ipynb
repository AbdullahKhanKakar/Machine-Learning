## One-Hot encoding using sklearn

```python
df.head(3)
```

```python
from sklearn.model_selection import train_test_split

x = df.drop("selling_price",axis=1)
y = df["selling_price"]
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.20, random_state=0)

x_train.shape, x_test.shape
```

```python
from sklearn.preprocessing import OneHotEncoder

en = OneHotEncoder(drop='first',sparse_output=False,dtype=np.int32)
en.fit(x_train[["fuel","owner"]])

x_train_new = en.transform(x_train[["fuel","owner"]])
x_test_new = en.transform(x_test[["fuel","owner"]])
```

```python
x_train_new
```

```python
x = x_train[["brand","km_driven"]].values
y = x_train_new

dataset = np.hstack((x,y))
dataset
```
